```python
from google.colab import files
uploaded = files.upload()
import io, pandas as pd
fn = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[fn]), sep=None, engine='python')
print(df.shape)
display(df.head(10))
print(df.columns.tolist())
print(df.dtypes)
print(df.isnull().sum())
```

```python
def clean_col(c):
    c = str(c).strip().lower()
    c = c.replace('(', '').replace(')', '')
    c = c.replace('?', '')
    c = c.replace('/', '_')
    c = c.replace('-', '_')
    c = c.replace(' ', '_')
    for ch in [',', ':', ';']:
        c = c.replace(ch, '')
    while '__' in c:
        c = c.replace('__', '_')
    return c

df.columns = [clean_col(c) for c in df.columns]
df.columns.tolist()
```

```python
df['date_parsed'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')

df['year'] = df['date_parsed'].dt.year
df['month'] = df['date_parsed'].dt.month
df['day'] = df['date_parsed'].dt.day
df['day_of_week'] = df['date_parsed'].dt.day_name()
df['is_weekend'] = df['day_of_week'].isin(['Saturday', 'Sunday'])

df[['date', 'date_parsed', 'year', 'month', 'day', 'day_of_week', 'is_weekend']].head(50)
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

checks = {}
for c in numeric_cols:
    checks[c] = {
        'min': float(df[c].min()),
        'max': float(df[c].max()),
        'zeros': int((df[c] == 0).sum()),
        'negatives': int((df[c] < 0).sum())
    }

checks
```

```python
conds = (df['hydraulic_pressurebar']<0) | (df['spindle_vibrationm']<0) | (df['spindle_speedrpm']==0) | (df['torquenm']==0)
print('rows_with_issues_count:', int(conds.sum()))
display(df.loc[conds, ['date_parsed','machine_id','assembly_line_no','hydraulic_pressurebar','spindle_vibrationm','spindle_speedrpm','torquenm','coolant_temperature','downtime']].head(50))
```

```python
df.loc[df['spindle_vibrationm'] < 0, 'spindle_vibrationm'] = 0
df = df[df['hydraulic_pressurebar'] >= 0].reset_index(drop=True)
df.shape
```

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
num_cols = [c for c in num_cols if c not in ['year','month','day']]

for col in num_cols:
    df[col] = df.groupby('machine_id')[col].transform(lambda x: x.fillna(x.median()))

df.isnull().sum()
```

```python
sensor_cols = [
    'hydraulic_pressurebar', 'coolant_pressurebar', 'air_system_pressurebar',
    'coolant_temperature', 'hydraulic_oil_temperaturec',
    'spindle_bearing_temperaturec', 'spindle_vibrationm',
    'tool_vibrationm', 'spindle_speedrpm', 'voltagevolts',
    'torquenm', 'cuttingkn'
]

outlier_counts = {}

for c in sensor_cols:
    q1 = df[c].quantile(0.25)
    q3 = df[c].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    df[c + '_is_outlier'] = ((df[c] < lower) | (df[c] > upper)).astype(int)

    df[c] = df[c].clip(lower, upper)

    outlier_counts[c] = int(df[c + '_is_outlier'].sum())

outlier_counts
```

```python
df['downtime_flag'] = df['downtime'].apply(
    lambda x: 1 if str(x).strip() == 'Machine_Failure' else 0)

df['downtime_flag'].value_counts()
```

```python
def infer_reason(row):
    if row['coolant_temperature'] > 70:
        return 'Overheating'
    if row['hydraulic_pressurebar'] < 60:
        return 'Hydraulic_Pressure_Drop'
    if row['spindle_vibrationm'] > 1.5:
        return 'High_Vibration'
    return 'Normal'

df['reason_preview'] = df.apply(infer_reason, axis=1)
df['reason_preview'].value_counts()
```

```python
df['temp_diff'] = df['coolant_temperature'] - df['hydraulic_oil_temperaturec']

df['vibration_total'] = df['spindle_vibrationm'] + df['tool_vibrationm']

df['pressure_ratio'] = df['hydraulic_pressurebar'] / df['coolant_pressurebar']

df['normalized_speed'] = df['spindle_speedrpm'] / df['spindle_speedrpm'].max()

df['load_index'] = (df['torquenm'] * 0.7) + (df['cuttingkn'] * 0.3)

df[['temp_diff','vibration_total','pressure_ratio','normalized_speed','load_index']].head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_data.csv")
df.head()
if "reason_preview" in df.columns:
    df = df.drop(columns=["reason_preview"])
```

```python
df.head()
```

```python
def get_reason(row):


    if row['hydraulic_pressurebar'] < 90 and row['coolant_temperature'] > 27:
        return 'High_CoolantTemp_with_Low_HydraulicPressure'

    if row['hydraulic_pressurebar'] < 90 and row['cuttingkn'] > 3.0:
        return 'High_CuttingForce_with_Low_HydraulicPressure'

    if row['hydraulic_pressurebar'] < 90 and row['torquenm'] < 24:
        return 'Low_Torque_with_Low_HydraulicPressure'

    if row['hydraulic_pressurebar'] < 90:
        return 'Low_HydraulicPressure'

    if row['cuttingkn'] > 3.0:
        return 'High_CuttingForce'

    if row['torquenm'] < 23:
        return 'Low_Torque'

    if row['coolant_temperature'] > 27:
        return 'High_CoolantTemperature'

    if row['load_index'] < 16:
        return 'Low_LoadIndex'

    return 'Normal_Operation'


df['final_reason'] = df.apply(get_reason, axis=1)

```

```python
df.to_csv("final_dataset_with_reason.csv", index=False)
```
